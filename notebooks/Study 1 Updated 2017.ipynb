```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('../plots/nyt_wsj_2017/nyt_2017.json') as results:
    data_nyt = json.load(results)
    
with open('../plots/nyt_wsj_2017/wsj_2017.json') as results:
    data_wsj = json.load(results)
```

```python
data = [data_nyt, data_wsj]
plt.figure()

fig, ax = plt.subplots()
ax.set_ylabel("Results per senator")
ax.set_title('Distribution of senator references, including outliers')

plt.boxplot(data, 
            labels=['New York Times', 'Wall Street Journal'])

# plt.savefig('../plots/nyt_wsj_2017/boxplot.png', dpi=250)
plt.show()
```

```python
plt.figure()

fig, ax = plt.subplots()
ax.set_title("I.Q. Scale")

plt.hlines(1,-1,1)  # Draw a horizontal line
plt.xlim(-1.01,1.01)
plt.ylim(0.5,1.5)

            

# Add tick for x=0 and remove the axes.
def adjust_spines(ax, spines):
    for loc, spine in ax.spines.items():
        if loc in spines:
            spine.set_position(('outward', 10))  # outward by 10 points
            spine.set_smart_bounds(True)
        else:
            spine.set_color('none')  # don't draw spine

    # turn off ticks where there is no spine
    if 'left' in spines:
        ax.yaxis.set_ticks_position('left')
    else:
        # no yaxis ticks
        ax.yaxis.set_ticks([])

    if 'bottom' in spines:
        ax.xaxis.set_ticks_position('bottom')
    else:
        # no xaxis ticks
        ax.xaxis.set_ticks([])
        
adjust_spines(ax, [])
plt.xticks([-1,0,1])
plt.gca().yaxis.set_major_locator(plt.NullLocator())
# plt.axis('off')
# plt.plot(0, np.ones(np.shape(0)), '|', ms=40, color='black')


scores = [0.06535, 0.241795] # [NYT, WSJ]
y = np.ones(np.shape(scores))   # Make all y values the same

plt.plot(scores, y, '.', ms = 20) # Plot points


ax.annotate('NYT 0.06', xy=(scores[0], 0.95), xytext=(scores[0], 0.75), ha='center',
            arrowprops=dict(facecolor='black', shrink=0.1, width=2)
            )
ax.annotate('WSJ 0.24', xy=(scores[1], 1.05), xytext=(scores[1], 1.25), ha='center',
            arrowprops=dict(facecolor='black', shrink=0.1, width=2)
            )
ax.annotate('Conservative', xy=(0.95, 0.95), ha='center', color='r')
ax.annotate('Liberal', xy=(-0.95, 0.95), ha='center', color='b')

# plt.savefig('../plots/nyt_wsj_2017/scale.png', dpi=250)
plt.show()
```
