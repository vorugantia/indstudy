Test `matplotlib`'s boxplot.

See demos here: https://matplotlib.org/examples/pylab_examples/boxplot_demo.html

```python
import json
import matplotlib.pyplot as plt
import numpy
```

```python
with open('../plots/senators_2016/nyt_senators_results.json') as results:
    data_nyt = json.load(results)
    
with open('../plots/senators_2016/wsj_senators_results.json') as results:
    data_wsj = json.load(results)
    
# print(data_nyt)
# print(data_wsj)
```

## NYT boxplots

```python
plt.boxplot(data_nyt)
plt.show()
```

```python
plt.figure()
plt.boxplot(data_nyt, 0, sym='') # Don't show outliers
plt.show()
```

## Combined NYT + WSJ boxplot

```python
data = [data_nyt, data_wsj]
plt.figure()
plt.boxplot(data)
plt.show()
```

```python
plt.figure()
plt.boxplot(data, 0, sym='') # Don't show outliers
plt.show()
```

## Transformed scale boxplots

Because the extreme outlier data points are so far from the box, let's try different scales: logarithmic and square root.

### 1. Logarithmic

```python
data = [numpy.log10(data_nyt), numpy.log10(data_wsj)]
plt.figure()
plt.boxplot(data)
plt.show()
```

### 2. Square root

```python
data = [numpy.sqrt(data_nyt), numpy.sqrt(data_wsj)]
plt.figure()
plt.boxplot(data)
plt.show()
```
