Test `matplotlib`'s boxplot.

See demos here: https://matplotlib.org/examples/pylab_examples/boxplot_demo.html

```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('../plots/senators_2016/nyt_senators_results.json') as results:
    data_nyt = json.load(results)
    
with open('../plots/senators_2016/wsj_senators_results.json') as results:
    data_wsj = json.load(results)
    
# print(data_nyt)
# print(data_wsj)
```

## NYT boxplots

```python
plt.boxplot(data_nyt)
plt.show()
```

```python
plt.figure()
plt.boxplot(data_nyt, 0, sym='') # Don't show outliers
plt.show()
```

## Combined NYT + WSJ boxplot

```python
data = [data_nyt, data_wsj]
plt.figure()

fig, ax = plt.subplots()
ax.set_ylabel("Results per senator")
ax.set_title('Distribution of senator references')

plt.boxplot(data, 
            labels=['New York Times', 'Wall Street Journal'])

# plt.savefig('../plots/senators_2016/boxplot.png', dpi=250)
plt.show()
```

```python
plt.figure()

fig, ax = plt.subplots()
ax.set_ylabel("Results per senator")
ax.set_title('Distribution of senator references, excluding outliers')

plt.boxplot(data, 
            0, sym='', #whis=??
            labels=['New York Times', 'Wall Street Journal']
           )

# plt.savefig('../plots/senators_2016/boxplot_exclusion.png', dpi=250)
plt.show()
```

## D/R bar graph of results

```python
N = 2
D_results = (934, 1816)
R_results = (1012, 3204)

ind = np.arange(N)
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(ind, D_results, width, color='b')
rects2 = ax.bar(ind + width, R_results, width, color='r')

ax.set_ylabel("Total references")
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('New York Times', 'Wall Street Journal'))
ax.set_title('Senator references by party, excluding outliers')

ax.legend((rects1[0], rects2[0]), ('D/I', 'R'))

# def autolabel(rects):
#     """
#     Attach a text label above each bar, displaying its value
#     """
#     for rect in rects:
#         height = rect.get_height()
#         ax.text = (rect.get_x() + rect.get_width()/2, 
#                    height*1.05, 
#                    'test',
#                    horizontalalignment='center',
#                    verticalalignment='bottom',
#                    color='black'
#                    )

# autolabel(rects1)
# autolabel(rects2)

# plt.savefig('../plots/senators_2016/bar_graph.png', dpi=250)
plt.show()
```
